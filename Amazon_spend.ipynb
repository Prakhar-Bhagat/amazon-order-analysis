```python
import pandas as pd
import numpy as np
df=pd.read_csv("/Users/prakharbhagat/Downloads/amazon-orders.csv")
df.head()
```

```python
df=df.fillna(0)
df.head()
```

```python
df["Total Charged"]=df["Total Charged"].str.replace('$','').astype(float)
df.head()
```

```python
df["Tax Charged"]=df["Tax Charged"].str.replace("$","").astype(float)
df.head()
```

```python
df["Total Charged Numeric"] = df["Total Charged"]
bins = np.linspace(df["Total Charged Numeric"].min(), df["Total Charged Numeric"].max(), 4)
group_names = ["low", "medium", "high"]
df["Total Charged Binned"] = pd.cut(df["Total Charged Numeric"], bins, labels=group_names, include_lowest=True)
```

```python
import matplotlib.pyplot as plt

df["Total Charged Binned"].value_counts().sort_index().plot(kind='bar', color='green', edgecolor='black')

plt.title('Spending Distribution by Category')
plt.xlabel('Spending Category')
plt.ylabel('Number of Orders')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```
